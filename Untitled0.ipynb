```python
def transform(data):
  events =["event_name_1","event_type_1","event_name_2","event_type_2"]
  for event in events:
    data[event].fillna("unknow",inplace=True)
  cat = ["event_name_1","event_type_1","event_name_2","event_type_2","snap_CA","snap_WI","snap_TX"]
  for event in cat:
    encoder = LabelEncoder()
    data[event] = encoder.fit_transform(data[event])
  return data 
```

getting ready for training

```python
X = []
for i in (range(time_series_data.shape[0])):
  X.append([list(t) for t in zip(transfer_calss.loc["event_name_1"][-(100+28):-(28)],
           transfer_calss.loc["event_type_1"][-(100+28):-(28)],
           transfer_calss.loc["event_name_2"][-(100+28):-(28)],
           transfer_calss.loc["event_type_2"][-(100+28):-(28)],
           transfer_calss.loc["snap_CA"][-(100+28):-(28)],
           transfer_calss.loc["snap_TX"][-(100+28):-(28)],
           transfer_calss.loc["snap_WI"][-(100+28):-(28)],
           price_df.iloc[i][-(100+28):-(28)],
           time_series_data.iloc[i][-100:])])
```

```python
X = np.asarray(X,dtype=np.float32)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import RepeatVector,LSTM,TimeDistributed,Dense
```

```python
model = Sequential()
model.add(LSTM(128,activation="relu",input_shape=(28,9),return_sequences=False))
model.add(RepeatVector(28))
model.add(LSTM(32,activation="relu",return_sequences=True))
model.add(TimeDistributed(Dense(1)))
```

```python
model.compile(optimizer="adam",loss="mse")
```

```python
def normilize(list):
  list = np.array(list)
  low,high = np.percentile(list,[0,100])
  dealta = high-low
  if dealta!=0:
    for i in range(0,len(list)):
      list[i] = (list[i]-low)/dealta
  return list,low,high         
```

```python
train_n,train_low,train_high = normilize(X)
x_train = train_n[:,-28*2:-28,:] #we use only 56 days data
y_train = train_n[:,-28:,8]
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1],9)
y_train = y_train.reshape(y_train.shape[0],y_train.shape[1],1)
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=10)
```

making predications
